# Оставим только необходимые колонки датасета

```python
columns_to_keep = [
    'NumberOfRooms',
    'GrossSquareMeters',
    'FloorLocation',
    'NumberFloorsofBuilding',
    'BuildStatus',
    'price',
    'district',
    'BuildingAge',
]

df_clean = df[columns_to_keep].copy()
```

```python
rename_map = {
    'NumberOfRooms': 'rooms',
    'GrossSquareMeters': 'square_gross',
    'FloorLocation': 'floor',
    'NumberFloorsofBuilding': 'total_floors',
    'BuildStatus': 'build_status',
    'price': 'price_tl',
    'district': 'district',
    'BuildingAge': 'building_age',
}

df_clean = df_clean.rename(columns=rename_map)
```
